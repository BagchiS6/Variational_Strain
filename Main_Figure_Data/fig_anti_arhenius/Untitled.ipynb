```python
import pandas as pd
def read_log(filename):
    with open(filename) as file:
        lines= file.readlines()
    with open('cache', 'w') as file:
        file.writelines(lines[198:198+5000])
    return pd.read_csv('cache', '\s+')

color = ['k', 'b', 'g', 'c', 'm', 'r']
ic= 0

for temp in [10, 150, 300, 350, 400, 500]:
    data = read_log(f'log.lammps.{temp}K')
    L0 = data.loc[0,'Lx']
    plt.plot((data.Lx-L0)/L0*100, data.PotEng - data.PotEng[0] +temp*100, color[ic], label=f'{temp}K seed1')
            
    
    if temp not in [10,350]:
            
                
            data_s2 = read_log(f'log.lammps.{temp}K.seed_2')
            L0 = data_s2.loc[0,'Lx']
            plt.plot((data_s2.Lx-L0)/L0*100, data_s2.PotEng - data_s2.PotEng[0] +temp*100, color[ic]+str('--'), label=f'{temp}K seed2')
            
            
            data_s3 = read_log(f'log.lammps.{temp}K.seed_3')
            L0 = data_s3.loc[0,'Lx']
            plt.plot((data_s3.Lx-L0)/L0*100, data_s3.PotEng - data_s3.PotEng[0] +temp*100, color[ic]+str('-.'), label=f'{temp}K seed3')
    ic+=1       
#     plt.xlim(2.82,2.85)
#     plt.ylim(-1.3318e8, -1.3317e8)
    plt.xlabel('strain (%)')
    plt.ylabel('Pot. Eng (eV)')
    plt.legend(loc=(1.1,-.1))
    
plt.savefig('strain_rate_Pot_energy_nucleation.png', bbox_inches='tight', dpi=600)
```
